```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\Y.PHANI MANJUNATH\\Downloads\\Dataset .csv")
df
```

```python
inputs=df.drop(['Restaurant ID','Restaurant Name','Address','Locality','Locality Verbose','Is delivering now','Switch to order menu'
                ,'Rating color', 'Rating text','Aggregate rating','City'],axis='columns')
inputs.head()
```

```python
from sklearn.preprocessing import LabelEncoder
letb=LabelEncoder()
leod=LabelEncoder()
lec=LabelEncoder()
```

```python
inputs['Has Table booking']=letb.fit_transform(inputs['Has Table booking'])
inputs['Has Online delivery']=leod.fit_transform(inputs['Has Online delivery'])
inputs['Currency']=lec.fit_transform(inputs['Currency'])
```

```python
inputs
```

```python
inputs['Cuisines']=inputs['Cuisines'].fillna('Other')
```

```python
inputs['Cuisines_list'] = inputs['Cuisines'].str.split(',').apply(lambda x: [c.strip() for c in x])
```

```python
inputs.columns.tolist()
```

```python
inputs.head()
```

```python
inputs['Primary_Cuisine'] = inputs['Cuisines_list'].apply(lambda x: x[0] if len(x) > 0 else 'Other')
```

```python
inputs.head()
```

```python
inputs.Primary_Cuisine.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
inputs['Primary_Cuisine']=le.fit_transform(inputs['Primary_Cuisine'])
```

```python
inputs.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cols_to_scale = ['Longitude', 'Latitude', 'Average Cost for two', 'Votes']
inputs[cols_to_scale] = scaler.fit_transform(inputs[cols_to_scale])
```

```python
inputs.head(10)
```

```python
y=inputs['Primary_Cuisine']
x=inputs.drop(['Cuisines','Cuisines_list','Primary_Cuisine'],axis='columns')
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=300,class_weight='balanced')
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
```
