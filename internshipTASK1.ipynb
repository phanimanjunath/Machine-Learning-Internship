```python
from collections import Counter
cuisine_counter=Counter(flatten_list)
print(cuisine_counter)
```

```python
print(cuisine_counter.most_common(10))
```

```python
top_k=10
topk_cuisines=[cuisine for cuisine,count in cuisine_counter.most_common(10)]
print(topk_cuisines)
```

```python
def filter_top_k(cuisine_list):
    return [cuisine if cuisine in topk_cuisines else 'Other' for cuisine in cuisine_list]

inputs['Cuisines_list'] = inputs['Cuisines_list'].apply(filter_top_k)
```
